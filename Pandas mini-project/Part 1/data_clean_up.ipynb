```python
# Dependencies
import pandas as pd
```

```python
# load CSV
df = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv")
```

```python
# Read with pandas
df.head()
```

```python
# Inspect all columns
df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)
columns = [0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
extract_df = df.iloc[:,columns]
extract_df.head(100)
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace function
extract_df.AttendedBootcamp.replace(to_replace=[0 ,1], value=['no', 'yes'])
extract_df.head(100)
```

```python
# Calculate total number of respondents in survey
# consider using the len() function
tot=len(extract_df["AttendedBootcamp"])
tot
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column
attended_df = extract_df.loc[extract_df["AttendedBootcamp"] == "yes"]

attended_df.head(20)
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it
age=round(attended_df["Age"].mean())
age
```

```python
# Calculate how many people attended a bootcamp
# Count the values of the AttendedBootcamp column
attendance=attended_df["AttendedBootcamp"].count()

attendance
```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column
sch_dg=attended_df["SchoolDegree"].value_counts()

sch_dg
```

```python
# Count the number of records where the person is a degree holder
# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column
bach = attended_df["AttendedBootcamp"].loc[attended_df["SchoolDegree"] == "bachelor's degree"].count()
MsD=attended_df["AttendedBootcamp"].loc[attended_df["SchoolDegree"] == "master's degree (non-professional)"].count()
prof=attended_df["AttendedBootcamp"].loc[attended_df["SchoolDegree"] == "professional degree (MBA, MD, JD, etc.)"].count()
assoc=attended_df["AttendedBootcamp"].loc[attended_df["SchoolDegree"] == "associate's degree"].count()
phd=attended_df["AttendedBootcamp"].loc[attended_df["SchoolDegree"] == "Ph.D."].count()
dg_h=bach+MsD+prof+assoc+phd
dg_h
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
attended_df=attended_df.replace({"genderqueer": "non-binary",
                               "agender": "non-binary", "trans":"non-binary"})
nw_gdr_count=attended_df["Gender"].value_counts()
nw_gdr_count
```

```python
# Calculate percentage of respondents who attended a bootcamp

pct_att=round((attendance/tot)*100,4)
pct_att
```

```python
# Calculate percentage of respondents belonging to each gender

m = attended_df["AttendedBootcamp"].loc[attended_df["Gender"] == "male"].count()
pct_m=round((m/sum(nw_gdr_count))*100,4)
f = attended_df["AttendedBootcamp"].loc[attended_df["Gender"] == "female"].count()
pct_f=round((f/sum(nw_gdr_count))*100,4)
nb = attended_df["AttendedBootcamp"].loc[attended_df["Gender"] == "non-binary"].count()
pct_nb=round((nb/sum(nw_gdr_count))*100,4)

print(pct_m)
print(pct_f)
print(pct_nb)
```

```python
# Calculate percentage of attendees with a college degree

pct_d=round((dg_h/attendance)*100,4)
pct_d
```

```python
# Calculate average post-bootcamp salary
avg=round(attended_df["BootcampPostSalary"].mean(),4)
avg
```

```python
# Create a new table consolidating above calculations
summary_df=pd.DataFrame([{"Total Surveyed":tot,
                          "Total Bootcamp attendees":attendance,
                          "Attended bootcamp (%)":pct_att,
                          "Average Age":age,
                          "Degree holder (%)":pct_d,
                          "% Male":pct_m,
                          "% Female":pct_f,
                          "% Non-binary":pct_nb,
                          "Average Post Bootcamp Salary":avg}])
                         
summary_df
```

```python
# Improve formatting before outputting spreadsheet

summary_df["Attended bootcamp (%)"] = summary_df["Attended bootcamp (%)"].map("%{:.2f}".format)
summary_df["Degree holder (%)"] = summary_df["Degree holder (%)"].map("%{:.2f}".format)
summary_df["% Male"] = summary_df["% Male"].map("%{:.2f}".format)
summary_df["% Female"] = summary_df["% Female"].map("%{:.2f}".format)
summary_df["% Non-binary"] = summary_df["% Non-binary"].map("%{:.2f}".format)
summary_df["Average Post Bootcamp Salary"] = summary_df["Average Post Bootcamp Salary"].map("${:.2f}".format)
summary_df
```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes

summary=summary_df.to_excel("Output/DataCleanSummary.xls", index=False, header=True)
```
