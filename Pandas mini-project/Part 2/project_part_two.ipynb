```python
# Dependencies
import pandas as pd
```

```python
# load CSV
df = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv")
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
# You may need to encode this file
df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Consider using iloc
extract_df = df.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
extract_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
extract_df=extract_df.replace([0 ,1], ['No', 'Yes'])
extract_df.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended_df = extract_df.loc[extract_df["AttendedBootcamp"] == "Yes"]
attended_df
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
# Create a DataFrame from using value_counts
bt_nm=attended_df["BootcampName"].value_counts()
bt_nm_df=pd.DataFrame(bt_nm)

bt_nm_df.reset_index(inplace=True)

bt_nm_df.columns=['BootcampName','Count']

bt_nm_df
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recomm = attended_df["BootcampName"].loc[extract_df["BootcampRecommend"] == "Yes"].value_counts()
recomm_df=pd.DataFrame(recomm)

recomm_df.reset_index(inplace=True)

recomm_df.columns=['BootcampName','Recommenders']

recomm_df

```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_df = pd.merge(bt_nm_df, recomm_df, on="BootcampName", how="right")
merge_df
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
# Think of what should be the numerator and the demoninator
merge_df['% Recommended']=round((merge_df['Recommenders']/merge_df['Count'])*100,2)

# Sort results in descending order
merge_df_s=merge_df.sort_values('% Recommended', ascending=False)
# Format for percentages

merge_df_s["% Recommended"] = merge_df_s["% Recommended"].map("{:.2f}%".format)
merge_df_s
```

```python
# Export to excel and remove index
merge=merge_df_s.to_excel("Output/DataCleanSummary.xls", index=False, header=True)
```
